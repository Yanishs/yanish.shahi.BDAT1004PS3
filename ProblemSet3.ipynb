#### Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort the actors by their kill count and label each bar with the corresponding actor's name. Don't forget to label your axes!

```python
#import necessary libraries 
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
#import emoji

#name the data frame impport from the csv file
actors_df = pd.read_csv("actor_kill_counts.csv")
```

```python
actors_df.head()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,8))

#plots the bar using seaborn
ax= sns.barplot(x="Count",y="Actor",data=sorted_df) #palette="viridis")

#for labels on each bars
for i, count in enumerate(sorted_df['Count']):
    
    ax.text(count +8,i, str(count)+" Kills")
    
    
plt.xlabel("kill count")
plt.ylabel("Actor")
plt.title("Dangerous Hollywood Actors")
plt.tight_layout()
plt.show()
```

# Question 9
#### Create a pie chart showing the fraction of all Roman Emperors that were assassinated.
#### Make sure that the pie chart is an even circle, labels the categories, and shows the |percentage breakdown of the categories.

```python
import pandas as pd
from matplotlib import pyplot as plt
#import seaborn as sns

emperors_df= pd.read_csv("roman-emperor-reigns.csv")
```

```python
emperors_df.head(10)
```

```python
#count the occurences of assasinated and non assasinated emperors

assassinate_emp = emperors_df[emperors_df["Cause_of_Death"]=="Assassinated"].count()["Cause_of_Death"]
non_assassinated_emp = emperors_df[emperors_df["Cause_of_Death"] != "Assassinated"].count()["Cause_of_Death"]

total_emperors = assassinate_emp+non_assassinated_emp

#percentage calculations

assasinated_emp_per = (assassinate_emp/total_emperors)*100
non_assassinated_emp_per = (non_assassinated_emp/total_emperors)*100

#data for pie chart
labels = "Assassinated Emperors", "Non-Assassinated emperors"
size = [assasinated_emp_per,non_assassinated_emp_per]
colors = ["red","green"]


#plotting chart
plt.figure(figsize=(8,8))
plt.pie(size,labels=labels, colors=colors, autopct="%1.2f%%", startangle=120)
plt.axis("equal")
plt.title("Fraction of Roman Emperors Assassinated")
plt.show()
```
